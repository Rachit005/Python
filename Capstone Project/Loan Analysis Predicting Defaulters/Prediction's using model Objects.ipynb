```python
# Importing the required libraries

import pandas as pd
from pickle import load
```

```python
# Loading the column names which are categorical
col = load(open('col.txt','rb'))

# Loading the unique values of the columns 
unique_values = load(open('uniquevalues.sav','rb'))

# Loading the Label Encoder Object
label = load(open('labelencoder.sav','rb'))

# Loading the Scaler Object
scale = load(open('scaler.sav','rb'))

# Loading the Model Object
model = load(open('model.sav','rb'))
```

```python
# Column Names which are categorical
col
```

```python
# Unique values of the categorical Columns

unique_values
```

```python
# Loading the data

data = pd.read_csv('New_processed_data.csv', index_col = 0)

l = ['default','issue_date','first_credit_year','issue_date_year','credit_report_year',
     'last_payment_date_year']
data1 = data.drop(l, axis = 1)
data1.head()
```

```python
data1[col].head()
```

```python
# Applying the Label Encoder

j = 0
for i in col:
    label.classes_ = unique_values[j]
    data1[i] = label.transform(data1[i])
    j += 1
```

```python
data1[col].head()
```

```python
# Scaling the data

data2 = scale.transform(data1)
data2
```

```python
model
```

```python
# Making the prediction using the Model Object

Y_pred = model.predict(data2)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print(confusion_matrix(data['default'],Y_pred))
print('\nClassification Report\n',classification_report(data['default'],Y_pred))
print('Accuracy of Model :',round(accuracy_score(data['default'],Y_pred),4))    
```
